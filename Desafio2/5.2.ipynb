```python
import numpy as np
import matplotlib.pyplot as plt
# np.random.seed(0)
n = 10000
R0= 1
Theta = np.random.uniform(0,2*np.pi,n)
T = np.random.gamma(3,1,n)
print(f"T : mean = {np.mean(T)}, var = {np.std(T)**2}")
X = R0*T*np.cos(Theta)
print(f"X : mean = {np.mean(X)}, var = {np.std(X)**2}")
Y = R0*T*np.sin(Theta)
print(f"Y : mean = {np.mean(Y)}, var = {np.std(Y)**2}")
plt.figure(figsize=(5,5))
plt.scatter(X,Y)
plt.show()
plt.hist(T,bins=20)
plt.figure(figsize=(5,5))
plt.show()
```
